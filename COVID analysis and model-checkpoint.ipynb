```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### CLEANING

```python
covid = pd.read_csv('Covid Data.csv') #load data
covid.head()
```

```python
covid = covid[covid['CLASIFFICATION_FINAL'].isin([1,2,3])] #including only confirmed COVID case
covid['DIED'] = covid['DATE_DIED'].apply(lambda x:0 if x == '9999-99-99' else 1) #binarizing to either died or survived
covid = covid.drop(columns=['DATE_DIED'])
covid
```

```python
binary_cols = ['SEX','PATIENT_TYPE','PNEUMONIA', 'PREGNANT','DIABETES','COPD','ASTHMA','INMSUPR',
               'HIPERTENSION','RENAL_CHRONIC', 'OTHER_DISEASE','TOBACCO','INTUBED','ICU']
for col in binary_cols: # fixes binary data, original data had 2 meaning false/no, and 97/98/99 representing missing data
    covid[col][covid[col] ==2] = 0
    covid[col][covid[col] ==97] = 0
    covid[col][covid[col] ==98] = 0
    covid[col][covid[col] ==99] = 0
```

```python
covid = covid.rename(columns={'SEX':'FEMALE','PATIENT_TYPE':'RETURNED_HOME','CLASIFFICATION_FINAL':'HAS_COVID'}) #name fixing
covid['HAS_COVID'] = 1 #each row has covid
```

# Model Building

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
covid.head(3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(covid.drop(columns=['DIED']), covid['DIED'], test_size=0.33, random_state=42)
```

### Baseline model

```python
#baseline score
#baseline is all 0 classification
print(f'test set:{accuracy_score(y_test, np.zeros(len(y_test)))}')
print(f'train set:{accuracy_score(y_train, np.zeros(len(y_train)))}')
```

```python
# baseline using logistic regression + top 4 differences in proportions in binary columns
from sklearn.linear_model import LogisticRegression
```

```python
logic_baseline = LogisticRegression().fit(X_train[['INTUBED','ICU','RETURNED_HOME','PNEUMONIA']], y_train)
```

```python
print(f"test set:{accuracy_score(y_test, logic_baseline.predict(X_test[['INTUBED','ICU','RETURNED_HOME','PNEUMONIA']]))}")
print(f"train set:{accuracy_score(y_train, logic_baseline.predict(X_train[['INTUBED','ICU','RETURNED_HOME','PNEUMONIA']]))}")
```

#### Naive bayes model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
#hyperparameter testing
for i in [10,1,0.1,0.0001]:
    bayes = MultinomialNB()
    bayes.fit(X_train, y_train)
    print(f'alpha={i},acc={accuracy_score(y_train, bayes.predict(X_train))}')
```

```python
#final
bayes = MultinomialNB()
bayes.fit(X_train, y_train)
print(f'test set:{accuracy_score(y_test, bayes.predict(X_test))}')
print(f'train set:{accuracy_score(y_train, bayes.predict(X_train))}')
```
